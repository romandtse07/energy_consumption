```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
```

```python
def tssplit(df, fraction = 0.7, max_lag=24):
    splindex = int(len(df)*fraction)
    #df = df.assign(diffd = df.consumption.diff())
    for i in [1,12,24]:
        df = df.join(df.consumption.shift(i).rename(f'lag{i}'))
    df.drop('temperature', axis=1, inplace=True)
    df.dropna(inplace=True)
    return df.iloc[:splindex, :], df.iloc[max_lag + splindex:, :]
```

```python
def prepare(df):
    scaler = MinMaxScaler(feature_range=(-1, 1))
    before, after = tssplit(df)
    
    lags = [column for column in before.columns if 'lag' in column]
    
    xtrain = scaler.fit_transform(before[lags])
    xtrain = xtrain.reshape((xtrain.shape[0],1,xtrain.shape[1]))

    xtest = scaler.transform(after[lags])
    xtest = xtest.reshape((xtest.shape[0],1,xtest.shape[1]))

    ytrain = before.consumption
    ytest = after.consumption
    return xtrain, xtest, ytrain, ytest
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
from keras.optimizers import Adam
from tqdm import tqdm
```

```python
model = Sequential([
    LSTM(units=3, batch_input_shape=(1,1,3), stateful=True),
    #Dense(64, activation='relu'),
    Dense(1)
])

model.compile(optimizer=Adam(), loss='mean_absolute_error')
```

```python
for i in range(3):
    i=0
    for idee in train.series_id.unique():
        xtrain, xtest, ytrain, ytest = prepare(train.query(f'series_id=={idee}'))

        model.fit(xtrain, ytrain, validation_data=(xtest, ytest), batch_size=1, verbose=0, shuffle=False)
        model.reset_states()
        if i%100==0:
            print(model.evaluate(xtest, ytest,batch_size=1))
        i+=1
    print('gotem')
```
