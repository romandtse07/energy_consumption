## LSTM Model

Most of the method of training is taken from [this blog post](http://drivendata.co/blog/benchmark-cold-start-lstm-deep-learning/) as a baseline with some adjustments.  We add some features such as surface size, base temperature, the hour, month and weekday.  In hindsight, the problem trained on is completely different from the problem at hand; there is no way to validate this form of the model, as it requires inputs by the hour which are constantly provided in the datasets here, but are missing for the test set.  I should probably stop while I'm ahead here.

```python
#ordinal probably means very little but oh well
surface_size = {
    'xx-small':-1,
    'x-small':-2/3,
    'small':-1/3,
    'medium':0,
    'large':1/3,
    'x-large':2/3,
    'xx-large':1
}
```

```python
train.surface = train.surface.map(lambda x: surface_size[x])
train.base_temperature = train.base_temperature.map(lambda x: 1 if x == 'low' else -1)
train.hour = train.hour/11.5 - 1
train.weekday = train.weekday/3 - 1
train.month = (train.month-1)/5.5 - 1
train.is_off = train.is_off*2 - 1
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
def tssplit(df, fraction, max_lag):
    splindex = int(len(df)*fraction)
    scaler = MinMaxScaler(feature_range=(-1, 1))
    scaler.fit(df[['consumption']].iloc[:splindex,:])
    df.consumption = scaler.transform(df[['consumption']])
    
    for i in range(max_lag):
        df = df.join(df.consumption.shift(i).rename(f'lag{i}'))
    df.dropna(inplace=True)
    return df.iloc[:splindex,:], df.iloc[splindex:, :]
```

```python
def prepare(df, fraction = 0.5, max_lag = 4):
    before, after = tssplit(df, fraction, max_lag)
    
    xtrain = before.drop(['consumption', 'series_id'], axis=1).values.reshape((before.shape[0], before.shape[1]-2, 1))
    xtest = after.drop(['consumption', 'series_id'], axis=1).values.reshape((after.shape[0], after.shape[1]-2, 1))

    ytrain = before.consumption.values
    ytest = after.consumption.values
    return xtrain, xtest, ytrain, ytest
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
from keras.optimizers import Adam
```

```python
feature_cols = ['series_id', 'consumption', 'surface', 'base_temperature', 'hour', 'weekday', 'month', 'is_off']
```

```python
max_lags = 4

model = Sequential([
    LSTM(units=20, 
         batch_input_shape=(1,len(feature_cols)+max_lags-2,1), 
         stateful=True, 
         activation='tanh', 
         dropout=0.2, 
         recurrent_dropout=0.2),
    Dense(1)
])

model.compile(optimizer=Adam(amsgrad=True), loss='mean_absolute_error')
```

```python
#not sure if this is a fair comparison to look at the loss of a holdout instead of training on it or something.
#definitely not looking at the loss the same way you would in the test set
for i in range(2):
    i=0
    for idee in train.series_id.unique():
        xtrain, xtest, ytrain, ytest = prepare(train[feature_cols].query(f'series_id=={idee}'), 0.8, max_lags)

        model.fit(xtrain, ytrain, validation_data=(xtest, ytest), batch_size=1, verbose=0, shuffle=False, epochs=1)
        model.reset_states()
        if i%100==0:
            print(model.evaluate(xtrain, ytrain, batch_size=1))
            print(model.evaluate(xtest, ytest,batch_size=1))
        i+=1
    print('gotem')
```
